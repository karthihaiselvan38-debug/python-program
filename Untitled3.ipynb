```python
import numpy as np

# Example: Sales data (Rows = Orders, Columns = [Sales, Quantity, Profit])
sales_data = np.array([
    [2500, 10, 200],
    [3000, 15, 350],
    [4000, 20, 500],
    [3500, 18, 400],
    [5000, 25, 800]
])

print("Sales Data:\n", sales_data)
```

```python
mean_all = np.mean(sales_data, axis=0)
median_all = np.median(sales_data, axis=0)
variance_all = np.var(sales_data, axis=0)
std_all = np.std(sales_data, axis=0)

print("Mean (Sales, Quantity, Profit):", mean_all)
print("Median:", median_all)
print("Variance:", variance_all)
print("Standard Deviation:", std_all)
```
